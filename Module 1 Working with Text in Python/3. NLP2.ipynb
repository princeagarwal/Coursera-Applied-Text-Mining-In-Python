## VID 2 Regular Expressions

```python
text = '"Ethics are built right into the \
idelas and objectives of the unitwed states \
into the UN " #UNSG @ NY Society for Ethical\
Cilture bit.ly/sfd @UN @UN_Women'
```

```python
text1 = text.split(" ")
```

```python
[w for w in text1 if w.startswith('@')]
```

```python
[w for w in text1 if w.startswith('#')]
```

```python
# Importing regular Expressions
import re
```

```python
[w for w in text1 if re.search('@[A-Za-z0-9_]+', w)]
```

```python
[w for w in text1 if re.search('@\w+', w)]
```

### finding Special char

```python
text2 = 'ouagadougou'

re.findall(r'[aeiou]', text2)
```

```python
re.findall(r'[^aeiou]', text2)
```

### Date variations

```python
datestr = '23-10-2002\n23/10/2002\n23/10/02\n10/23/2002\n23 Oct 2202\n23 October 2002\nOct 23,2002\nOctober 23, 2002\n'
```

```python
re.findall(r'\d{2}[/-]\d{2}[/-]\d{4}', datestr)
```

```python
re.findall(r'\d{2}[/-]\d{2}[/-]\d{2,4}', datestr)
```

```python
re.findall(r'\d{2} (Jan|Feb|Mar|Apr|May|Jun|jul|Aug|Sep|Oct|Nov|Dec) \d{4}', datestr)
```

```python
re.findall(r'\d{2} (?:Jan|Feb|Mar|Apr|May|Jun|jul|Aug|Sep|Oct|Nov|Dec) \d{4}', datestr)
```

```python
re.findall(r'\d{2} (?:Jan|Feb|Mar|Apr|May|Jun|jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4}', datestr)
```
